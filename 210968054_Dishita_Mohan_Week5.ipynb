```python
# 16) Find the Total runs in each season

DF=pd.read_csv("deliveries.csv")
match_detail=df[['id', 'season']].merge(DF, left_on='id', right_on='match_id', how='left')

match_detail.groupby(['season'])['total_runs'].sum().reset_index()
```

```python
# 18) Visualize the Toss decision across seasons

ax = plt.axes()
ax.set(facecolor = "white")

sns.countplot(x='season', hue='toss_decision', data=df,palette="gnuplot2",saturation=1)
# plt.xticks(rotation=90,fontsize=10)
# plt.yticks(fontsize=10)
plt.xlabel('\n Season',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title('Toss decision in different seasons',fontsize=15,fontweight="bold")
plt.show()
```

```python
# 20) Find the Top 10 run scorers in IPL and Visualize using best fit graph

top10=match_detail.groupby(['batsman'])['total_runs'].sum().reset_index()

top10.sort_values(by=['total_runs'], ascending=False).head(10)

plt.hist(top10)
plt.show()
```
